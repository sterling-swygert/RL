```python
df = df.drop(columns=['zipcode'])
```

```python
# Many zeroes for year renovated... Means they just werent renovated.
# But it turns out this is okay, because the more recently a home was renovated,
# so zero actually makes a lot of sense. Alternatively, we could fill in 
# with yr_build value if yr_renovated is zero, or just create a boolean 'is_renovated'
# but we will just leave it as is as it makes sense

df['yr_renovated'].value_counts()
```

```python
# similar thing with sqft_basement, i.e., no basement is worse than small basement

df['sqft_basement'].value_counts()
```

```python
X = df.drop(columns=['price']).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Do scaling post split to do it only to training and avoid data leakage

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)

#X_eval, X_test, y_eval, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=101)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(X_train.shape[1], activation='relu'))
model.add(Dense(X_train.shape[1], activation='relu'))
model.add(Dense(X_train.shape[1], activation='relu'))
model.add(Dense(X_train.shape[1], activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test),
         batch_size=128, epochs=400)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df
```

```python
loss_df.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_pred, y_test)
mae = mean_absolute_error(y_pred, y_test)
```

```python
print(f'Average price: ${round(y.mean())}.')
print(f'RMSE: ${round(np.sqrt(mse))}')
print(f'MAE: ${round(mae)}')
```

```python
# We are off by about 20% in MAE... not very good. Let's check out the variance

explained_variance_score(y_test, y_pred)
```

```python
# Our model is actually pretty good below $3,000,000, so maybe we can do better if we exclude the outlying high
# priced homes


plt.figure(figsize=(12,6))
plt.scatter(y_test, y_pred)
plt.ticklabel_format(style='plain', axis='both')
plt.plot(y_test, y_test, 'r')
```
